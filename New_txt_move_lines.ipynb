```python
#explore the size and content of each line in a text file

folderpath = '/Users/carsonlam/macroeyes/population1/'
txtname = 'cpt'
i=0

for MRN in dirs:
    
    MRN = MRN.rstrip()
    
    textpath = folderpath + MRN + '/' + txtname + '.txt'
    
    if os.path.exists(textpath):

        f = open(textpath, "r",encoding = 'iso-8859-1') 

        readfile = csv.reader(f)

        filelist = list(readfile)

        f.close()
    
    cpt = []
    icd9 = []
    
    for line in filelist[1:]:
        
        if len(line) > 5:
            newline = line[:3] + line[-2:]
            combo = ','.join(line[2:-2])
            newline[2] = combo
            line = newline
            
        if line[3] == 'CPT':
            #cpt.append(line)
            cpt.append(line)
            
        elif line[3] == 'ICD9CM':
            #icd9.append(line)
            icd9.append(line)
            
    with open(folderpath + MRN + '/' + 'Cur_Proc_Term' + '.txt', "w") as myfile:
        myfile.write("%s\n" % 'PAT_ID,PROC_CODE,PROC_NAME,PROC_CODE_1,CODE_DATE')
        for item in cpt:
            item[0] =str(item[0])
            s=','
            item = s.join(item)
            myfile.write("%s\n" % item)
            
    with open(folderpath + MRN + '/' + 'proc_icd' + '.txt', "w") as myfile:
        myfile.write("%s\n" % 'PAT_ID,PROC_CODE,PROC_NAME,PROC_CODE_1,CODE_DATE')
        for item in icd9:
            item[0] =str(item[0])
            s=','
            item = s.join(item)
            myfile.write("%s\n" % item)
    i+=1
    if i % 1000 == 0:
        print(i/len(dirs))
```

```python
#explore the size and content of each line in a text file

folderpath = '/Users/carsonlam/macroeyes/population1/'
txtname = 'VA_IOP'
i=0

for MRN in dirs:
    
    MRN = MRN.rstrip()
    
    textpath = folderpath + MRN + '/' + txtname + '.txt'
    
    if os.path.exists(textpath):

        f = open(textpath, "r",encoding = 'iso-8859-1') 

        readfile = csv.reader(f)

        filelist = list(readfile)

        f.close()
    
    VA = []
    IOP = []
    
    allrows = filelist[1:]
    
    for row in allrows:
        
        if 'VA' in row[2]:
            VA.append(row)
            
        if 'pressure' in row[2] or 'Tonometry' in row[2]:
            IOP.append(row)

    with open(folderpath + MRN + '/' + 'VA' + '.txt', "w") as myfile:
        myfile.write("%s\n" % 'PAT_MRN_ID, CONTACT_DATE, EXAM_CODE, EXAM_VALUE')
        for item in VA:
            item[0] =str(item[0])
            s=','
            item = s.join(item)
            myfile.write("%s\n" % item)
            
    with open(folderpath + MRN + '/' + 'IOP' + '.txt', "w") as myfile:
        myfile.write("%s\n" % 'PAT_MRN_ID, CONTACT_DATE, EXAM_CODE, EXAM_VALUE')
        for item in IOP:
            item[0] =str(item[0])
            s=','
            item = s.join(item)
            myfile.write("%s\n" % item)
            
    i+=1 
    
    if i % 1000 == 0:
        print(i/len(dirs))
```
