```python
import os
import glob
import csv
import copy
from functions import *
```

```python
folderpath = '/Users/carsonlam/macroeyes/population1/'
dirs = get_immediate_subdirectories(folderpath)
len(dirs)
```

```python
txtname = 'IOP'

i = 0

for MRN in dirs:
    
    MRN = MRN.rstrip()
    
    textpath = folderpath + MRN + '/' + txtname + '.txt'
    
    if os.path.exists(textpath):

        f = open(textpath, "r",encoding = 'iso-8859-1') 

        readfile = csv.reader(f)

        allrows = list(readfile)

        f.close()
        
        header = allrows[0]
        if len(header) == 4:
            #print(header)
            header.append('Tonometry_Method') 
            header.append('Tonometry_Time') 
        #print(header,len(header))
        datarows = allrows[1:]
        #print(len(datarows))
        lastrow = ['nothing']
        newrows = []
        
        s=','
        for row in datarows:
            if row != lastrow:
                newrows.append(row)
            lastrow = row

        if  os.path.exists(textpath):
            file = open(textpath, "w")
            header = s.join(header)
            file.write(header + "\n")
            file.close()             

        with open(textpath, "a") as myfile:
            for item in newrows:
                item = s.join(item)
                myfile.write("%s\n" % item)
        
        i+=1

        if i%100 == 0:
            print(i/len(dirs))
```
