```python
txtname = 'VA'

i = 0

for MRN in dirs:
    
    MRN = MRN.rstrip()
    
    textpath = folderpath + MRN + '/' + txtname + '.txt'
    
    if os.path.exists(textpath):

        f = open(textpath, "r",encoding = 'iso-8859-1') 

        readfile = csv.reader(f)

        allrows = list(readfile)

        f.close()
        
        header = allrows[0]
        
        datarows = allrows[1:]
        
        s=','
        file = open(textpath, "w")
        header = s.join(header)
        file.write(header + "\n")
        file.close() 
        #print(header)
        
        correction_value = ' '
        method = 0
        va_value = []
        if len(datarows) > 1:
            
            last_date_value = datarows[0][1]
        
            for row in datarows:

                if row[1] != last_date_value:
                    with open(textpath, "a") as myfile:
                        for item in va_value:
                            item.append(method_type)
                            item.append(correction_value)
                            item = s.join(item)
                            myfile.write("%s\n" % item)
                            #print(item)

                    va_value = []    
                    method = 0
                    correction_value = ' '
                    #print('Move to new date, TIME + METHOD =',(time+method),'IOP=',iop)

                if ('Right' in row[2]) or ('Left' in row[2]):
                    va_value.append(row)

                if 'Correction' in row[2]:
                    correction_value = row[3]

                if 'Method' in row[2]:
                    method += 1

                    if method > 1:
                        with open(textpath, "a") as myfile:
                            for item in va_value:
                                item.append(method_type)
                                item.append(correction_value)
                                item = s.join(item)
                                myfile.write("%s\n" % item)
                                #print(item)

                        va_value = []    
                        method = 0
                        correction_value = ' '

                    method_type = row[3]

                last_date_value = row[1]
            
            with open(textpath, "a") as myfile:
                for item in va_value:
                    item.append(method_type)
                    item.append(correction_value)
                    item = s.join(item)
                    myfile.write("%s\n" % item)
                    #print(item)
            method = 0
            va_value = []  
            correction_value = ' '
            
        i+=1
        #if i > 0:
            #break  
        
        if i%100 == 0:
            print(i/len(dirs))
```
