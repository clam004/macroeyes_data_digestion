```python
import os
import glob
import csv
import copy

csvname = 'VA_IOP'

folderpath = '/Users/carsonlam/macroeyes/population1/' 

datapath = '/Users/carsonlam/macroeyes/2nd_extraction/'

csvpath = datapath + csvname + '.csv'

f = open(csvpath, "rt", encoding = 'iso-8859-1') 

readfile = csv.reader(f)

filelist = list(readfile)

#title = filelist[0]
#notitle = filelist[1:]
#length = len(notitle)
length = len(filelist)

#s=','
#cleantitle = s.join(title)
print(length)
```

```python
i = 0

lastcleanrow ='-'

for row in filelist:  
    
    ptMRN = row[1].rstrip()
    date = row[2]
    value = row[6].rstrip()
    exam_name = row[7].rstrip()
    #print(ptMRN)
    #['PAT_MRN_ID', 'PAT_ENC_CSN_ID', 'CONTACT_DATE', 'EXAM_CODE', 'EXAM_VALUE']
    textpath = folderpath+str(ptMRN)+'/'+csvname+'.txt'
    
    #cleanrow = s.join(row)
    cleantitle = 'PAT_MRN_ID, CONTACT_DATE, EXAM_CODE, EXAM_VALUE'
    cleanrow = ptMRN + ',' + date + ',' + exam_name + ',' + value 
    lastcleanrow = copy.deepcopy(cleanrow)
    
    if not os.path.exists(folderpath+str(ptMRN)):
        os.makedirs(folderpath+str(ptMRN))
    
    if not os.path.exists(textpath):
        #print('make',textpath )
        file = open(textpath, "w")
        #print(cleantitle)
        file.write(cleantitle+ "\n")
        file.close()
        
    if cleanrow != lastcleanrow:
        with open(textpath, "a") as myfile:
            myfile.write(cleanrow + "\n")
    
    i+=1
    
    if i%10000 == 0:
        print(i/length)
```
